```python
print('WELCOME TO CALCI.CALCULATOR\n')
import math
class basic_calci:
    
    def summ(self):
        a=int(input('enter 1st :'))
        b=int(input('enter 2nd :'))
        print('Sum of Two number is : ',a+b)
    def subs(self):
        a=int(input('enter 1st :'))
        b=int(input('enter 2nd :'))
        print('Substraction of number is : ',a-b)
    def multy(self):
        a=int(input('enter 1st :'))
        b=int(input('enter 2nd :'))
        print('Multiplication of given number is : ',a*b)
    def div(self):
        a=int(input('enter 1st :'))
        b=int(input('enter 2nd :'))
        print('Division is :',a/b)
    def mod(self):
        a=int(input('Dividend :'))
        b=int(input('Divisor : '))
        print('Remainder is : ',a%b)
# class for advance calculator and inheritance of basic calculator.
class ad_calci(basic_calci):
    
    def power(self):
        a=int(input('enter number :'))
        b=int(input('power :'))
        print('Power of given number is','(',a,'^',b,')',':',math.pow(a,b))
    def expo(self):
        e=int(input('number : '))
        print('Exponential of number is : ',math.exp(e))
    def sqr(self):
        s=lambda a:a**2
        a=int(input('Enter the number : ')) 
        print('Square of above number is : ',s(a))
    def sqrt(self):
        r=int(input('Root of : '))
        print('root is :',math.sqrt(r))
    def sin(self):
        a=int(input('sin of ?'))
        print('sin','(',a,')','is',math.sin(a))
    def cos(self):
        a=int(input('cos of ?'))
        print('cos','(',a,')','is',math.cos(a))
    def tan(self):
        a=int(input('tan of ?'))
        print('tan','(',a,')','is',math.tan(a))
    def sinh(self):
        a=int(input('sinh of ?' ))
        print('sinh','(',a,')','is',math.sinh(a))
    def cosh(self):
        a=int(input('cosh of ?'))
        print('cosh','(',a,')','is',math.cosh(a))
    def tanh(self):
        a=int(input('tanh of ?'))
        print('tanh','(',a,')','is',math.tanh(a))
    def factorial(self):
        f=int(input('Factorial of : '))
        print('Factorial of given number is : ',math.factorial(f))
# calculate is object where for above class        
calculate=ad_calci()

print('For basic Caculation\n','-'*50)
ch_bas={'sum':1,'Substraction':2,'Multiplication':3,'Division':4,'Remainder':5}
print(ch_bas,'\n')

print('For Advance Caculation\n','-'*50)
ch_ad={'Power':6,'exponential':7,'Square':8,'Square-root':9,'sin':10,'cos':11,'tan':12}
ch_ins={'sinh':13,'cosh':14,'tanh':15,'Factorial':16}
print(ch_ad)
print(ch_ins)

while True:
    choose=int(input('Select given Category from above :'))
    if choose==1:
        print(calculate.summ())
    elif choose==2:
        print(calculate.subs())
    elif choose==3:
        print(calculate.multy())
    elif choose==4:
        print(calculate.div())
    elif choose==5:
        print(calculate.mod())
    elif choose==6:
        print(calculate.power())
    elif choose==7:
        print(calculate.expo())
    elif choose==8:
        print(calculate.sqr())
    elif choose==9:
        print(calculate.sqrt())
    elif choose==10:
        print(calculate.sin())
    elif choose==11:
        print(calculate.cos())
    elif choose==12:
        print(calculate.tan())
    elif choose==13:
        print(calculate.sinh())
    elif choose==14:
        print(calculate.cosh())
    elif choose==15:
        print(calculate.tanh())
    elif choose==16:
        print(calculate.factorial())
    sug=input('You want to continue/ exit press [y/n] : ')
    if sug=='y':
        continue
    else:
        print('Thank You')
        break
        
```
